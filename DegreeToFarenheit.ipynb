This notebook shows how to use Google Colab for writing python code, executing it. We'll write a function for converting the temperature from degree celcius to degree farenheit.

```
# Import Dependencies
import numpy as np
import tensorflow as tf
# Enabler Eager Execution mode
tf.enable_eager_execution()
```

```
# Set up Training Data
celcius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
farenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```
celcius
```

```
farenheit
```

```
for i,c in enumerate(celcius):
  print("{0} degree Cecius = {1} degree Fahrenhet". format(c, farenheit[i]))
```

```
# Define the Layer for the Model
# input_shape = [1]; gets one value at a time from the array
# units = 1; number of neurons in the layer
l = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```
# Assemble layers into Model
model = tf.keras.Sequential([l])
```

```
# Compile the Model with Loss and Optimizer
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

```
# Get Model Summary
model.summary()
```

```
# Train the Model
history = model.fit(celcius, farenheit, epochs=600, verbose=True)
print('Finished Training the Model...')
```

```
# Display Training Statistics

# Import Dependencies
import matplotlib.pyplot as plt

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(history.history['loss'])
```

```
# Make Predictions using the Trained Model
prediction = model.predict([32.0])
print('{0} degree Celcius = {1} degree Farenhet.'.format("32.0", prediction[0][0]))
```

32 degrees Celcius is equal to 89.6 degrees in Farenhet. So, the prediction of the model is pretty close to the actual value.

```
# Layer Weights
print("Layer Variables: ", l.get_weights())
```

These values are close to 1.8 and 32 which are the actual values for the equation for celcius to farenhet conversion i.e.
F = 1.8 * C + 32

```
# Model-2
l0 = tf.keras.layers.Dense(units=4, input_shape=[1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=1)
model_2 = tf.keras.Sequential([l0, l1, l2])

# Compile the Model with Loss and Optimizer
model_2.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))

# Get Model Summary
model_2.summary()
```

```
# Train the Model
history_2 = model_2.fit(celcius, farenheit, epochs=600, verbose=True)
print('Finished Training the Model...')
```

```
# Display Training Statistics

# Import Dependencies
import matplotlib.pyplot as plt

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(history_2.history['loss'])
```

```
# Make Predictions using the Trained Model
prediction_2 = model_2.predict([32.0])
print('{0} degree Celcius = {1} degree Farenhet.'.format("32.0", prediction_2[0][0]))
```

```
# Layer Weight and Bias Values
print("Layer-0 Variables: {}\n".format(l0.get_weights()))
print("Layer-1 Variables: {}\n".format(l1.get_weights()))
print("Layer-2 Variables: {}\n".format(l2.get_weights()))
```
